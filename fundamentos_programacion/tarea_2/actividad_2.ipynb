```ruby
## EJERCICIO 1

def determinar_obsequio(precio, cantidad)
  
    vPremio = "Ninguno"
    vMonto  = (precio * cantidad).round(2)

    # programar a partir de aqui
  
    if (60..110).include? vMonto 
      vPremio = "Llavero"
    elsif (111..180).include? vMonto 
      vPremio = "Agenda"
    elsif vMonto > 180
      vPremio = "Casaca"      
    end

    return vPremio

end
    
def test_determinar_obsequio(regalo, precio, cantidad)

  print validate(regalo, determinar_obsequio(precio,cantidad))

end
    
def validate (expected, value)
 expected == value ? "." : "F"
end

def test
 
  puts "Test de prueba del programa"
  puts "---------------------------"
  test_determinar_obsequio("Ninguno", 5, 7)
  test_determinar_obsequio("Llavero", 7, 10)
  test_determinar_obsequio('Agenda' , 8, 15)
  test_determinar_obsequio( 'Casaca', 9, 22)

  puts " "
end

test
```

```ruby
## EJERCICIO 2

def determinar_monto_a_pagar(tipo_producto, tipo_celular, cantidad)
  
    vPrecio = 0.0
    vMonto  = 0.0

  # programar a partir de aqui
  
  if tipo_producto == 'O'
    if tipo_celular == 'Android'
      vPrecio = 20.30
    else
      vPrecio = 50.60
    end
  elsif tipo_producto == 'J'
    if tipo_celular == 'Android'
      vPrecio = 40.50
    else
      vPrecio = 90.80
    end    
  elsif tipo_producto == 'U'
    if tipo_celular == 'Android'
      vPrecio = 30.60
    else
      vPrecio = 60.50
    end        
  end
  
  vMonto = vPrecio * cantidad
  return vMonto
  
end
    
def test_determinar_monto_a_pagar(importe, tipo_producto, tipo_celular, cantidad)

  print validate(importe, determinar_monto_a_pagar(tipo_producto, tipo_celular, cantidad))

end
    
def validate (expected, value)
 expected == value ? "." : "F"
end

def test
 
  puts "Test de prueba del programa"
  puts "---------------------------"
  test_determinar_monto_a_pagar(1012, 'O', 'iPhone',  20)
  test_determinar_monto_a_pagar(1215, 'J', 'Android', 30)
  test_determinar_monto_a_pagar(2420, 'U', 'iPhone',  40)

  puts " "
end

test
```

```ruby
## EJERCICIO 3

def determinar_monto_a_vender(fabricante, tipo_Programa, dispositivo, cantidad)
  
    vPrecio = 0.0
    vMonto  = 0.0

    # a partir de aqui iniciar la programación
  
  if fabricante == 'Microsoft'
    if tipo_Programa == 'Sistema Operativo'
      vPrecio = 1500
    else
      if dispositivo == 'Celulares'
        vPrecio = 1800
      elsif dispositivo == 'Tablas'
        vPrecio = 1200
      else
        vPrecio = 900
      end
    end
  elsif fabricante == 'Apple'
    if tipo_Programa == 'Sistema Operativo'
      vPrecio = 2500
    else
      if dispositivo == 'Celulares'
        vPrecio = 1900
      elsif dispositivo == 'Tablas'
        vPrecio = 1800
      else
        vPrecio = 1600
      end
    end
  elsif fabricante == 'Linux'
    if tipo_Programa == 'Sistema Operativo'
      vPrecio = 1000
    else
      if dispositivo == 'Celulares'
        vPrecio = 100
      elsif dispositivo == 'Tablas'
        vPrecio = 150
      else
        vPrecio = 50
      end
    end  
  end
  
  vMonto = vPrecio * cantidad
  return vMonto
  
end
    
def test_determinar_monto_a_vender(importe, fabricante, tipo_Programa, dispositivo, cantidad)

  print validate(importe, determinar_monto_a_vender(fabricante, tipo_Programa, dispositivo, cantidad))

end
    
def validate (expected, value)
 expected == value ? "." : "F"
end

def test
 
  puts "Test de prueba del programa"
  puts "---------------------------"
  test_determinar_monto_a_vender(1800, 'Microsoft', 'Lenguaje de Programación',  'Otros', 2)
  test_determinar_monto_a_vender(7500, 'Apple'    , 'Sistema Operativo'       ,       '', 3)
  test_determinar_monto_a_vender( 300, 'Linux'    , 'Lenguaje de Programación', 'Tablas', 2)

  puts " "
end

test
```
